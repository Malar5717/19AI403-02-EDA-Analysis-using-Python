```python
import pandas as pd

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('titanic_dataset.csv')
df.head()
```

Replace null values using mean, median, or mode

```python
df.isnull().sum()
```

```python
df['Age'].fillna(value=df['Age'].median(), inplace=True)
df.head()
```

```python
df['Embarked'].fillna(value=df['Embarked'].mode()[0], inplace=True)
df.head()
```

```python
df['Cabin'].fillna('unknown', inplace=True)
df.head()
```

```python
df.isnull().sum()
```

Boxplot to analyze outliers

```python
# visualizes the distribution, median, and outliers of numerical data

plt.figure(figsize=(10,5))
sns.boxplot(x=df['Fare'])
# potential skewing, meaningless values
plt.title('Boxplot of Fare')
plt.show()
```

Remove outliers using IQR method

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5* IQR
upper_bound = Q3 + 1.5* IQR

df_clean = df[(df['Fare'] >= lower_bound) & (df['Fare'] <= upper_bound)]
df_clean.head()
```

Use Countplot to analyze categorical data

```python
# shows counts of categories

plt.figure(figsize=(10,5))
sns.countplot(x=df['Survived'])
plt.title('Survival count')
plt.show()

plt.figure(figsize=(10,5))
sns.countplot(x=df['Sex'], palette={'male': '#4A90E2', 'female': '#FF69B4'})
plt.title('Gender distribution')
plt.show()
```

Use displot to show univariate distribution

```python
# shows distribution of a single variable
# i. kind='kde' - smooth curve

sns.displot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

# ii. kind='hist' - frequency distribution
# iii. kind='ecdf' - proportion of observations  each value
```

Use cross tabulation

```python
cross_tab = pd.crosstab(df['Survived'], df['Sex'])
print(cross_tab)

cross_tab.plot(kind='bar', stacked=False)
plt.title('Survived vs Sex')
plt.show()
```

Use heatmap to show variable relationships

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
